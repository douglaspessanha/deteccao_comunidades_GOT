### Tarefa 1 - Discuta quais as principais diferenças entre ambas as redes com base na distribuição de graus dos vértices (apresente os gráficos de histograma), grau médio, densidade e transitividade. Para estas três medidas, apresente os valores obtidos durante a discussão.

```python
# carregando o csv da primeira temporada e transformando num Grafo
S1 = pd.read_csv("/home/douglas/onedrive/PUC_disciplinas/tecnicas_em_graph_mining/atividade_somativa_2/got-s1-edges_semanas7_8.csv")

G1=nx.from_pandas_edgelist(S1, source="Source", target="Target", edge_attr="Weight")
```

```python
# carregando o csv da oitava temporada e transformando num Grafo
S8 = pd.read_csv("/home/douglas/onedrive/PUC_disciplinas/tecnicas_em_graph_mining/atividade_somativa_2/got-s8-edges_semanas7_8.csv")

G8=nx.from_pandas_edgelist(S8, source="Source", target="Target", edge_attr="Weight")
```

```python
# grau médio da primeira temporada
graus1 = dict(G1.degree())
graus1 = list(graus1.values())
grau_medio = np.mean(graus1)
grau_medio
```

```python
# grau médio da oitava temporada
graus8 = dict(G8.degree())
graus8 = list(graus8.values())
grau_medio = np.mean(graus8)
grau_medio
```

```python
# distribuição dos graus da primeira temporada
plt.xlabel("Grau")
plt.ylabel("Distribuição")
plt.hist(graus1)
```

```python
# distribuição dos graus da oitava temporada
plt.xlabel("Grau")
plt.ylabel("Distribuição")
plt.hist(graus8)
```

```python
# densidade da primeira temporada
nx.density(G1)
```

```python
# densidade da oitava temporada
nx.density(G8)
```

```python
#agrupamento global na primeira temporada
nx.transitivity(G1)
```

```python
#agrupamento global na oitava temporada
nx.transitivity(G8)
```

### Tarefa 2) Para cada uma das redes, identifique quem são os 3 personagens mais centrais de acordo com as medidas de centralidade de grau, intermediação, proximidade e autovetor. Algum destes personagens é central em ambas as temporadas?

```python
#centralidade de autovetor temporada primeira
autovetor1 = nx.eigenvector_centrality(G1)
for i in sorted(autovetor1, key = autovetor1.get, reverse = True):
    print(i, autovetor1[i])
```

```python
#centralidade de promimidade temporada primeira
proximidade1 = nx.closeness_centrality(G1)
for i in sorted(proximidade1, key = proximidade1.get, reverse = True):
    print(i, proximidade1[i])
```
